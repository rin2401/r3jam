```python

def solve_single(t, arr):
    c = 360 * 12 * 1e10
    for i in range(3):
        found = False
        for j in range(3):
            if i == j: continue
            gs = arr[i]
            gm = arr[j]
            gh = arr[3-i-j]    
            for h in range(12):
                n = ((3600*1e9*h - (gh-gm))/11 + c) % c
                if not n.is_integer(): continue
                ns = int(n % 1e9)
                s = int(n//1e9 % 60)
                m = int(n//1e9//60%60)
                dh = int(n//1e9//60//60)
                h = (h + dh) % 12 
                ggs = (1e9 * s + ns) * 60 * 12
                ggm = m * 6 * 12 * 1e10 + (1e9 * s + ns) * 12
#                 print("Case #{}: {} {} {} {}".format(t + 1, h, m, s, ns))
#                 print((gm - gs + c) % c , (ggm - ggs + c) % c)
                if (gm - gs + c) % c == (ggm - ggs + c) % c:
                    print("Case #{}: {} {} {} {}".format(t + 1, h, m, s, ns))
                    return

def solve():
    T = int(input())
    for t in range(T):
        arr = list(map(int, input().split()))
        solve_single(t, arr)
```
